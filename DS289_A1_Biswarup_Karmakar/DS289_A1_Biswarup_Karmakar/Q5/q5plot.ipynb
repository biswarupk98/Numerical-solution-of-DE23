```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

s_list = ['100','500', '1000', '1500']

for s in s_list:
    temp_csv = pd.read_csv('radius_temp_' + s + '.txt', sep = '\t')
    plt.figure()
    plt.plot(temp_csv.iloc[:, 0], temp_csv.iloc[:, 1])
    plt.title('Radius vs Temperature  Plot for s = ' + str(s) + '')
    plt.xlabel('Radius')
    plt.ylabel('Temperature')
    plt.show()
    
    
plt.figure()
for s in s_list:
    temp_csv = pd.read_csv('radius_temp_' +s + '.txt', sep = '\t')
    plt.plot(temp_csv.iloc[:, 0], temp_csv.iloc[:, 1], label = 'S = ' + s)
plt.title('Radius vs Temperature  Plot for different S')
plt.xlabel('Radius')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```
