```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
c = '5'
dt_list = ['2.500000','1.000000', '0.500000', '0.250000', '0.100000']
method_name = 'time_displacement_' + c + '_'
in_csv=pd.read_csv('time_displacement_5_0.100000.txt',sep='\t')
plt.figure()
for i in dt_list:
    temp_csv = pd.read_csv(method_name + i + '.txt', sep = '\t')
    plt.plot(temp_csv.iloc[:, 0], temp_csv.iloc[:, 1], label = 'stepsize = ' + i)
plt.plot(in_csv.iloc[:, 0], in_csv.iloc[:, 3],linestyle='dashed', label = 'analytical solution')
plt.title('Displacement vs Time Curves for Different step sizes')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.legend()
plt.show()


plt.figure() 
for i in dt_list:
    error_csv = pd.read_csv(method_name + i + '.txt', sep = '\t')
    plt.plot( np.log10(error_csv.iloc[:, 0]),np.log10(error_csv.iloc[:, 2]), label='stepsize = ' + i)

plt.title('Time vs Error curves in log scale for different step size')
plt.xlabel('log(Time)')
plt.ylabel('log(Error)')
plt.legend()
plt.show()

optimal_h = '0.100000'
c_list = ['5', '40', '200']


plt.figure()
for i in c_list:
    temp_csv = pd.read_csv('time_displacement_' + i + '_' + optimal_h + '.txt', sep = '\t')
    plt.plot(temp_csv.iloc[:, 0], temp_csv.iloc[:, 1], label = 'c = ' + i)

plt.title('Displacement vs Time Curves for Different damping coefficients')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.legend()
plt.show()

```
